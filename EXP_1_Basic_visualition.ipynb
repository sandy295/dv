```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
big_mac = pd.read_csv('/content/big-mac-source-data.csv')
big_mac.head(10)

with open('/content/big-mac-source-data.csv', 'r') as file:
   data = file.read()
big_mac_sorted = big_mac.sort_values(by='GDP_dollar', ascending=False)
top_10_gdp_countries = big_mac_sorted.head(10)
top_10_gdp_countries.head(10)
fig = px.bar(top_10_gdp_countries, x='GDP_dollar', y='name', orientation='h', color='GDP_dollar',
labels={'GDP_dollar': 'GDP per person ($)', 'name': 'Country'}, title='Top 10 Countries based on GDP per person (July 2019)')
fig.update_layout(xaxis_title='GDP per person ($)', yaxis_title='Country')
fig.show()

plt.figure(figsize=(10, 6))
plt.barh(top_10_gdp_countries['name'], top_10_gdp_countries['GDP_dollar'], color='skyblue')
plt.xlabel('GDP per person ($)')
plt.ylabel('Country')
plt.title('Top 10 Countries based on GDP per person (July 2019)')
plt.gca().invert_yaxis() # Invert y-axis to have the highest GDP on top
plt.show()


```

```python
import pandas as pd
import plotly.express as px
big_mac = pd.read_csv('/content/big-mac-source-data.csv')
big_mac_july_2019 = big_mac[big_mac['date'] == '2019-07-09']
big_mac_july_2019['dollar_price'] = big_mac_july_2019['local_price'] / big_mac_july_2019['dollar_ex']
big_mac_july_2019['dollar_ppp'] = big_mac_july_2019['local_price'] / big_mac_july_2019['dollar_price']
countries = big_mac_july_2019['name']
dollar_price = big_mac_july_2019['dollar_price']
dollar_ppp = big_mac_july_2019['dollar_ppp']
plt.figure(figsize=(12, 6))
bar_width = 0.35
plt.bar(countries, dollar_price, color='blue', width=bar_width, label='Dollar Price')
plt.bar(countries, dollar_ppp, color='green', width=bar_width, label='Dollar PPP', alpha=0.5)
plt.xlabel('Countries')
plt.ylabel('Value')
plt.title('Comparison of Dollar Price and Dollar PPP for All Countries (July 2019)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
